list와 tuple의 주된 차이점 은 무엇인가요?

## list와 tuple의 주된 차이점

* list와 tuple은 값을 변경할 수 있는가의 여부에 따라 나뉘는데
  프로그램이 실행 되는 동안 값이 변경되면 안 되는 경우는 튜플을 사용
  값을 변경해야 하는 경우에는 리스트를 사용할 수 있다.

[문제2]

Python의 set에 대해 설명해 주세요

## Python의 set에 대한 설명

* set는 집합에 관련된 것들을 쉽게 처리하기 위해서 만든 자료형이다.
  set는 중복을 허용하지 않는 특징이 있기 때문에 데이터의 중복을
  제거하기 위한 필터로 종종 사용이 된다.

[문제3]
Python의 list는 연결리스트(linked list) 인가요?

## Python의 list는 연결리스트(linked list) 인가요?

* 우선 연결리스트란 선형 자료구조 중에 가장 기본이 되는 개념으로 흩뿌려진 노드들이
  서로 연결된 형태로 구성된 자료구조를 의미한다.
  파이썬의 경우에는 동적배열로 구현이 되어있는 프로그램으로 독립적인 메모리 공간에
  저장되고, 각 요소가 다음 요소를 가르키는 포인터를 포함한 구조이다.

* 딕셔너리 자료형은 리스트나 튜플처럼 순차적으로 요소의 값을 구하지 않고
  Key를 통해서 Value를 얻는다. 
  딕셔너리를 생성하는 방법은 두 가지가 있는데, 중괄호{}를 사용하는 방법과
  내장함수 dict()를 사용하는 방법이 있다.
