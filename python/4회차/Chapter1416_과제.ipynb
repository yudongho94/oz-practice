#### 프로젝트: 붕어빵 자동 판매 시스템


당신은 붕어빵 가게의 주인입니다. 

손님들이 붕어빵을 주문하면 이를 자동으로 처리하고, 재고를 관리하는 시스템을 만들고자 합니다. 또한, 관리자 모드를 통해 재고를 추가할 수 있는 기능을 구현해야 합니다.


[문제 1: 붕어빵 재고 확인]

붕어빵 가게의 현재 재고를 확인하는 기능을 구현하세요.

요구사항:
- 현재 붕어빵의 재고를 딕셔너리로 관리합니다.
- 딕셔너리에는 붕어빵의 종류와 각각의 재고 개수가 저장됩니다.
- 현재 재고를 화면에 출력하세요.

붕어빵 재고
- 팥붕어빵 : 10개
- 슈크림붕어빵 : 8개
- 초코붕어빵 : 5개

```python
재고 = {"팥붕어빵":10, "슈크림붕어빵":8,"초코붕어빵":5}
for k, v in 재고.items():
  print(f'{k} : {v}')
```

[문제 2: 붕어빵 주문 받기]

손님이 주문한 붕어빵의 종류와 개수를 입력받는 기능을 구현하세요.

요구사항:
- 손님이 주문할 붕어빵의 종류와 개수를 입력받습니다.
- 주문 내용을 딕셔너리에 저장하세요.
- 주문 내역을 출력해주세요.

```python
주문_내역 = {}

붕어빵_종류 = input("팥붕어빵, 슈크림붕어빵,초코붕어빵중 구매하실 붕어빵을 입력해주세요: ")

붕어빵_개수 = int(input("개수를 선택해주세요: "))

주문_내역[붕어빵_종류] = 붕어빵_개수

print(f'주문 내역: {주문_내역}개')
```

[문제 3: 붕어빵 주문 처리]

손님의 주문을 처리하여 붕어빵 재고를 업데이트하는 기능을 구현하세요.

요구사항:
- 손님의 주문 내용을 기반으로 재고를 업데이트합니다.
- 재고가 부족할 경우, 손님에게 재고 부족을 알리고 재고를 감소시키지 않습니다.
- 재고가 충분할 경우, 주문한 만큼 재고를 감소시키고 판매를 완료합니다.
- 판매가 완료된 경우 판매된 붕어빵 맛과 개수를 출력하세요.

```python
for k, v in 주문_내역.items():
  if 재고[k]<v:
    print("재고가 부족합니다.")
  
  elif v != 0:
    재고[k] -= v
    print(f'판매 완료 {k}, {v}')
```

[문제 4: 여러 손님 주문 처리]

여러 손님들의 주문을 처리할 수 있도록 시스템을 확장하세요.

요구사항:
- 손님이 "종료"를 입력할 때까지 주문을 계속 받습니다.
- 각 주문마다 재고를 업데이트합니다.
- 현재 재고를 화면에 출력합니다.

```python
while True:
  붕어빵_종류 = input("팥붕어빵, 슈크림붕어빵,초코붕어빵중 구매하실 붕어빵을 입력해주세요: ")
  if 붕어빵_종류 == "종료":
    break
  붕어빵_개수 = int(input("개수를 선택해주세요: "))

  if 재고[붕어빵_종류] >= 붕어빵_개수:
    재고[붕어빵_종류] -= 붕어빵_개수
    print(f"{붕어빵_종류}를 {붕어빵_개수} 판매하였습니다.")
  else:
    print(f"죄송합니다. {붕어빵_종류}의 재고가 부족합니다.")
    print(f"현재 재고는{재고}입니다.")
```

[문제 5: 관리자 모드]

관리자 모드를 통해 붕어빵 재고를 추가할 수 있는 기능을 구현하세요.

요구사항:
- 관리자는 특정 붕어빵의 재고를 추가할 수 있습니다.
- "종료"를 입력하면 관리자 모드를 종료합니다.
- 추가된 재고를 업데이트하고 현재 재고를 화면에 출력합니다.

```python
while True:
  붕어빵_종류 = input("팥붕어빵, 슈크림붕어빵,초코붕어빵중 구매하실 붕어빵을 입력해주세요: . 종료를 원하시면 '종료'를 눌러주세요")
  if 붕어빵_종류 == "종료":
    break
  추가_개수 = int(input(f"{붕어빵_종류}추가할 개수를 입력해주세요: "))

  재고[붕어빵_종류] += 추가_개수

  print(f"{붕어빵_종류}의 재고가 {추가_개수}개 추가되었습니다.")
  print(f"현재 재고는 {재고}입니다.")
```
