[문제6]

ZeroDivisionError,  IndexError 외에 3가지 이상의 파이썬 에러에 대해 설명해주세요

## ValueError

* ValueError는 부적절한 값을 가진 인자를 받았을 때 발생하는 에러이다.
  예를 들어 int를 받아야하는 함수에 str을 받았을 경우 등등이 포함된다.

## NameError

* 지역변수, 전역변수 이름을 찾을 수 없을 경우 생성되는 NameError이다.
  위에서 선언하지 않는 변수를 사용하려고 할 때 발생하는 에러이다.

## FileNotFoundError

* 파일이나 디렉터리에 접근하려고 할 때, 해당 파일이나 디렉터리가 없는 경우에
  발생하는 에러이다.
